```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path = "Monte-Carlo.csv"
data = pd.read_csv(file_path, sep=';')
data = data.iloc[:, :-1]


print(data.head())


x = data['Num of gen points']


y_constant = 0.25 * np.pi + 1.25 * np.arcsin(0.8) - 1
print(f"Добавляемая линия y = {y_constant}")


plt.figure(figsize=(10, 10))

for column in data.columns[1:3]:
    plt.plot(x, data[column], label=column)

plt.axhline(y=y_constant, color='green', linestyle='--', label="Correct answer")


plt.xlim(500, 100000)
plt.ylim(0.85, 1.06)
plt.xlabel('Num of gen points')
plt.ylabel('Values')
plt.title('График зависимости от числа генерированных точек')
plt.legend()
plt.grid()

plt.savefig("output_graph.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path = "/content/IntroSort vs QuickSort.csv"
data = pd.read_csv(file_path, sep=',')


print(data.head())

x = data['ARRAY SIZE']
plt.figure(figsize=(20, 20))

for column in data.columns[1], data.columns[4]:
    plt.plot(x, data[column], label=column)



plt.xlim(500, max(x))
plt.ylim(0.0, 125000.0)
plt.xlabel('Array size')
plt.ylabel('Time')
plt.title('График зависимости времени выполнения от размера массива в микросекундах')
plt.legend()
plt.grid()


plt.savefig("randoms.png", dpi=300)
plt.show()
```
